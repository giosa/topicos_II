```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
celcius = np.array([-50, -28, -12, -3, 9, 18, 40],dtype=float)
fahrenheit = np.array([-58, -18.4, 10.4, 26.6, 48.2, 64.4, 104],dtype=float)
```

```python
capa = tf.keras.layers.Dense(units=1, input_shape=[1])
modelo = tf.keras.Sequential([capa])
```

```python
modelo.compile(
    optimizer=keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
print("we begin with training season is open")
historial= modelo.fit(celcius, fahrenheit, epochs=450,verbose=False)
print("modelo entrenado")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de perdida")
plt.plot(historial.history["loss"])
```

```python
print("vamoave como predice esta vaina")
resultado = modelo.predict(np.array([[10.0]]))
print("el resultado es " + str(resultado) + "fahrenheit")
```

```python
print("vamoave otra predicción de esta vaina")
resultado_2 = modelo.predict(np.array([[100.0]]))
print("el resultado es " + str(resultado_2) + "fahrenheit")
```
